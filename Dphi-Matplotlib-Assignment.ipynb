# Task - 2

```python
#a
#top 5 east asian countries based on generosity
east_asian = df1[df1['Regional indicator']=='East Asia'].sort_values(by='Generosity',ascending=False).head()
east_asian[['Country name','Generosity']].plot(kind='bar').set_xlabel('Country')
```

```python
#b
#top 5 east asian countries ladder score
plt.figure(figsize=(10,5))
plt.plot(east_asian['Country name'],east_asian['Ladder score'])
```

```python
#c
#top 5 south asian countries ladder score
south_asian = df1[df1['Regional indicator']=='South Asia'].sort_values(by='Generosity',ascending=False).head()
south_asian[['Country name','Generosity']].plot(kind='bar').set_xlabel('Country')
```
