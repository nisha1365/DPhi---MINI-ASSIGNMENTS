```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/world-happiness-report-2021.csv")
```

```python
df.head()
```

```python
df['Regional indicator'].unique()
```

```python
df_new = df[(df['Regional indicator']=='East Asia')|(df['Regional indicator']=='Southeast Asia')].sort_values(by='Generosity',ascending=False) 
df_new.head()
```

```python
df1 = df_new[df['Regional indicator']=='East Asia'].head()
plt.figure(figsize=(15, 6), dpi=80)
plt.plot(df1['Country name'],df1['Ladder score'])
plt.title('variation of "Ladder score" among the above 5 east asian countries')
plt.xlabel('Country name')
plt.ylabel('Ladder score')
plt.show()
```

```python
df2 = df_new[df['Regional indicator']=='Southeast Asia'].head()
plt.figure(figsize=(15, 6), dpi=80)
plt.plot(df2['Country name'],df2['Ladder score'])
plt.title('variation of "Ladder score" among the above 5 south east asian countries')
plt.xlabel('Country name')
plt.ylabel('Ladder score')
plt.show()
```
